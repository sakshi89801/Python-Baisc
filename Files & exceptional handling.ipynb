Q1. What is the difference between interpreted and compiled languages?


Interpreted and compiled languages differ in how code is executed.

Interpreted Languages:
Code is executed line by line.
No separate compilation step.
Errors are shown during execution.
Slower execution compared to compiled languages.
Easier to debug.

Example: Python, JavaScript.
Compiled Languages:
Entire code is converted into machine code at once.
Execution is faster.
Errors are shown after compilation.
Requires compiler before running program.

Example: C, C++.
Key Difference:
Interpreted = step-by-step execution
Compiled = whole program converted first, then executed

6. What is the difference between import and from ... import in Python?



In Python, modules can be imported in different ways depending on how functions are used.

1. import statement:
Imports the entire module.
We need to use module name to access functions.

Example:

import math
math.sqrt(25)
Advantages:
Avoids name conflicts
Clear code structure
2. from ... import statement:
Imports specific functions/classes from a module.
No need to write module name.

Example:

from math import sqrt
sqrt(25)
Advantages:
Code becomes shorter
Easier to use specific functions
Key Difference:
import module → full module access
from module import function → specific access

Q15. How does Python's garbage collection system work?


Python’s garbage collection system automatically manages memory by removing unused objects.

Key Mechanism:
1. Reference Counting:
Each object has a reference count.
When count becomes zero, object is deleted.
2. Garbage Collector (GC):
Handles circular references (objects referencing each other).
Runs automatically in background.

Working:
Creates objects in memory.
Tracks references.
Deletes unused objects.

Example of circular reference:
Object A references B
Object B references A
GC removes them when unused

Importance:
Prevents memory leaks
Efficient memory usage
Automatic memory cleanup

Q16. What is the purpose of the else block in exception handling?

Answer:

The else block in Python exception handling is used to execute code when no exception occurs in the try block.

Key Points:
Executes only if try block is successful.
Runs before finally block.
Helps separate normal execution logic from error handling.
Syntax:
try:
    # risky code

    # error handling
else:
    # runs if no error
finally:
    # always runs
Example:

    x = 10 / 2
except ZeroDivisionError:
    print("Error occurred")
else:
    print("Division successful")
Importance:
Improves code clarity
Separates error-free logic from exception handling
