1 What is Object-Oriented Programming (OOP)?

Object-Oriented Programming (OOP) is a programming paradigm based on the concept of objects and classes.
It organizes code into reusable and modular structures.
It focuses on real-world entities like objects (e.g., car, student, bank account).
It improves code reusability, maintainability, and scalability.
Main pillars: Encapsulation, Inheritance, Polymorphism, Abstraction.


2. What is a class in OOP?

A class is a blueprint or template for creating objects.
It defines properties (variables) and behaviors (methods).
Example: A “Car” class may include color, model, and speed.
It does not occupy memory until objects are created.
It helps in organizing data and functions together.


3. What is an object in OOP?


An object is an instance of a class.
It represents a real-world entity with state and behavior.
It has attributes (data) and methods (functions).
Example: A specific car like “Toyota Camry” is an object.
Objects occupy memory when created.


4. What is the difference between abstraction and encapsulation?


Abstraction:
Hides internal implementation details and shows only functionality.
Focuses on “what it does”.
Encapsulation:
Hides data and restricts direct access using access modifiers.
Focuses on “how data is protected”.
Abstraction is achieved using abstract classes.
Encapsulation is achieved using private variables and methods.
Example: ATM (user sees options, not internal logic).

5. What are dunder methods in Python?



Dunder methods are special methods surrounded by double underscores (__).
Example: __init__, __str__, __repr__.
They are also called magic methods.
Used to define behavior of objects for built-in operations.
Example: __add__ defines addition behavior.


6. Explain the concept of inheritance in OOP.



Inheritance allows a class to acquire properties of another class.
Parent class is called base class; child class is derived class.
Promotes code reusability.
Types: single, multiple, multilevel inheritance.
Example: Vehicle → Car inherits features.


7. What is polymorphism in OOP?


Polymorphism means “many forms”.
Same function name behaves differently in different situations.
Achieved through method overriding and overloading.
Improves flexibility and scalability.
Example: “+” operator works for numbers and strings.


8. How is encapsulation achieved in Python?


Encapsulation is achieved using private variables.
Prefix _var (protected) and __var (private).
Access controlled through getter and setter methods.
Data is hidden from direct access.
Improves security and control of data.


9. What is a constructor in Python?



A constructor is a special method __init__.
It is automatically called when an object is created.
Used to initialize object attributes.
Helps in setting initial values.
Example: def __init__(self, name):


10. What are class and static methods in Python?


Class method:
Uses @classmethod
Takes cls as parameter
Works with class variables
Static method:
Uses @staticmethod
Does not take self or cls
Works independently of class and object
Both belong to class but behave differently.


11. What is method overloading in Python?



Method overloading means multiple methods with same name but different parameters.
Python does not support true method overloading.
Achieved using default arguments or variable arguments.
Improves flexibility.
Example: def add(a, b=0)


12. What is method overriding in OOP?

Method overriding occurs when child class redefines parent class method.
Same method name and same parameters.
Used in inheritance.
Provides specific implementation in child class.
Example: Animal → Dog overrides sound().


13. What is a property decorator in Python?


property is used to make a method behave like an attribute.
Used for getter methods.
Helps in controlled access to private variables.
Improves encapsulation.
Example: property def name(self):


14. Why is polymorphism important in OOP?


It allows one interface to be used for different data types.
Improves code flexibility and reusability.
Reduces complexity.
Supports dynamic method binding.
Makes code easier to extend.


15. What is an abstract class in Python?


An abstract class cannot be instantiated.
It contains abstract methods (no implementation).
Defined using abc module.
Forces child classes to implement methods.
Used for designing templates.


16. What are the advantages of OOP?


Code reusability through inheritance.
Better organization using classes.
Data security using encapsulation.
Flexibility through polymorphism.
Easy maintenance and scalability.


17. What is multiple inheritance in Python?



Multiple inheritance allows a class to inherit from more than one class.
Combines features of multiple parent classes.
Example: class C(A, B)
Can lead to ambiguity (diamond problem).
Python resolves it using Method Resolution Order (MRO).


18. What is the difference between a class variable and an instance variable?



Class variable:
Shared among all objects
Defined inside class
Instance variable:
Unique for each object
Defined using self
Class variable = shared memory
Instance variable = object-specific data

```python
#2
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        return math.pi * self.r * self.r

class Rectangle(Shape):
    def __init__(self, l, b):
        self.l = l
        self.b = b

    def area(self):
        return self.l * self.b

c = Circle(5)
r = Rectangle(4, 6)

print("Circle Area:", c.area())
print("Rectangle Area:", r.area())
```

```python
#3
class Vehicle:
    def __init__(self, type):
        self.type = type

class Car(Vehicle):
    def __init__(self, type, brand):
        super().__init__(type)
        self.brand = brand

class ElectricCar(Car):
    def __init__(self, type, brand, battery):
        super().__init__(type, brand)
        self.battery = battery

ec = ElectricCar("Car", "Tesla", "100kWh")

print(ec.type)
print(ec.brand)
print(ec.battery)
```

```python
#4
class Bird:
    def fly(self):
        print("Some birds can fly")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies high")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")

b1 = Sparrow()
b2 = Penguin()

b1.fly()
b2.fly()
```

```python
#5
class BankAccount:
    def __init__(self):
        self.__balance = 0   # private variable

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient balance")

    def check_balance(self):
        return self.__balance

acc = BankAccount()
acc.deposit(1000)
acc.withdraw(300)
print(acc.check_balance())
```

```python
#6
class Instrument:
    def play(self):
        print("Instrument is playing")

class Guitar(Instrument):
    def play(self):
        print("Guitar is playing melodious sound")

class Piano(Instrument):
    def play(self):
        print("Piano is playing classical music")

i1 = Guitar()
i2 = Piano()

i1.play()
i2.play()
```

```python
#7
class MathOperations:

    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

print(MathOperations.add_numbers(10, 5))
print(MathOperations.subtract_numbers(10, 5))
```

```python
#8
class Person:
    count = 0

    def __init__(self):
        Person.count += 1

    @classmethod
    def total_persons(cls):
        return cls.count

p1 = Person()
p2 = Person()
p3 = Person()

print("Total Persons:", Person.total_persons())
```

```python
#9
class Fraction:
    def __init__(self, num, den):
        self.num = num
        self.den = den

    def __str__(self):
        return f"{self.num}/{self.den}"

f = Fraction(3, 4)
print(f)
```

```python
#10
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)

v3 = v1 + v2
print(v3)
```

```python
#12
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def average_grade(self):
        return sum(self.grades) / len(self.grades)

s = Student("Riya", [80, 90, 85])
print("Average:", s.average_grade())
```

```python
#13
class Rectangle:
    def set_dimensions(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area(self):
        return self.length * self.breadth

r = Rectangle()
r.set_dimensions(5, 4)
print("Area:", r.area())
```

```python
#14
class Employee:
    def __init__(self, hours, rate):
        self.hours = hours
        self.rate = rate

    def calculate_salary(self):
        return self.hours * self.rate

class Manager(Employee):
    def __init__(self, hours, rate, bonus):
        super().__init__(hours, rate)
        self.bonus = bonus

    def calculate_salary(self):
        return super().calculate_salary() + self.bonus

m = Manager(40, 500, 2000)
print("Salary:", m.calculate_salary())
```

```python
#15
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        return self.price * self.quantity

p = Product("Laptop", 50000, 2)
print("Total Price:", p.total_price())
```

```python
#16
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        print("Moo")

class Sheep(Animal):
    def sound(self):
        print("Baa")

c = Cow()
s = Sheep()

c.sound()
s.sound()
```

```python
#17
class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year = year_published

    def get_book_info(self):
        return f"Title: {self.title}, Author: {self.author}, Year: {self.year}"

b = Book("Python Basics", "John Doe", 2023)
print(b.get_book_info())
```

```python
#18
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

m = Mansion("Delhi", 5000000, 10)

print(m.address)
print(m.price)
print(m.number_of_rooms)
```
